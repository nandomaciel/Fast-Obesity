```
df_obesity = pd.read_csv("original_weight_data.csv")
```

```
df_obesity.head(2)
```

```
df_obesity.info()
```

```
df_obesity = df_obesity[df_obesity['Question'] == "Percent of adults aged 18 years and older who have obesity"]
```

```
df_obesity = df_obesity[df_obesity['YearEnd'] == 2016]
```

```
# Dividir coluna Geolocalization em latitude e longitude
df_obesity["Latitude"] = (df_obesity["GeoLocation"].str.split(n=1,expand=True).iloc[:,0])
df_obesity["Longitude"] = (df_obesity["GeoLocation"].str.split(n=1,expand=True).iloc[:,1])
df_obesity['Latitude'] = df_obesity['Latitude'].str.replace("(","")
df_obesity['Latitude'] = df_obesity['Latitude'].str.replace(",","")
df_obesity['Longitude'] = df_obesity['Longitude'].str.replace(")","")

obesityTotal = df_obesity[df_obesity['Total'] == "Total"]
obesityTotal = obesityTotal[['LocationAbbr', 'LocationDesc', 'Data_Value', 'Latitude', 'Longitude']]
obesityTotal = obesityTotal[obesityTotal['LocationDesc'] != 'National']

#obesityTotal['Latitude'] = obesityTotal['Latitude'].astype(float)
#obesityTotal['Longitude'] = obesityTotal['Longitude'].astype(float)

obesityTotal = obesityTotal.sort_values(by=['Data_Value'])


obesityTotal
```

```
#lista de listas contendo latitude, longitude e porcetagem de obesos por estado
geoLoc = zip(obesityTotal['Latitude'].astype(float),obesityTotal['Longitude'].astype(float),obesityTotal['Data_Value'])
#geoLoc
```

```
m = folium.Map(location =[40.730610,-73.935242],zoom_start =5,tiles = 'Stamen Terrain',width='75%' ,height='75%')

for i in range(54):
  folium.Marker([geoLoc[i][0],geoLoc[i][1]], popup=str(geoLoc[i][2]),tooltip = 'Clique!').add_to(m)
  
m
```
